```python
import matplotlib.colors as col
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import colormaps, cm
from mpl_toolkits.axisartist.axislines import AxesZero
# Putting axes from https://matplotlib.org/stable/gallery/axisartist/demo_axisline_style.html
```

```python
# 1.5 Characterize the YAG bar
# current vs voltage
data = [[1, -11.9],
        [15, -11.8],
        [25, -11.8],
        [35, -11.6],
        [40, -11.5],
        [45, -11.4],
        [50, -11.3],
        [55, -11.1],
        [60, -10.4],
        [65, -5.5],
        [70, 5],
        [75, 22],
        [80, 39],
        [120, 103]
]

data_ = np.array(data)
x = data_[:,0] # current in mA
y = data_[:,1] # voltage in mV

# Plotting
# Design from https://matplotlib.org/stable/gallery/axisartist/demo_axisline_style.html
fig = plt.figure()
ax = fig.add_subplot(axes_class = AxesZero)

for direction in ["xzero", "yzero"]:
    ax.axis[direction].set_axisline_style("-|>")
    ax.axis[direction].set_visible(False)

for direction in ["left", "bottom"]:
    ax.axis[direction].set_axisline_style("-|>")
    ax.axis[direction].set_visible(True)

for direction in ["right", "top"]:
    ax.axis[direction].set_visible(False)

ax.plot(x, y, marker = "o")
ax.set_xlabel("Current, mA")
ax.set_ylabel("Voltage, mV")
plt.grid(linestyle = "--")
plt.savefig("1_5_voltage_vs_current.jpg")
plt.show()
```

```python
# Exercise 1.5
# temperature vs. transmission
data2 = np.array([[7.2, 17],
[8.6, 24],
[10.8, 33],
[12.5, 36.5],
[14.3, 37],
[16, 39.5],
[17.8, 44.5],
[19.9, 46.5],
[21.5, 43.5],
[23.5, 36.5],
[25.4, 28.5],
[27.2, 17],
[28.8, 10],
[31.0, -3],
[33.0, -4],
[34.8, -2],
[36.4, -1],
[38.1, 0],
[40, -1.5],
[41.9, -4],
[44, -6.5]
])

fig = plt.figure()
ax = fig.add_subplot(axes_class = AxesZero)

for direction in ["xzero", "yzero"]:
    ax.axis[direction].set_axisline_style("-|>")
    ax.axis[direction].set_visible(False)

for direction in ["left", "bottom"]:
    ax.axis[direction].set_axisline_style("-|>")
    ax.axis[direction].set_visible(True)

for direction in ["right", "top"]:
    ax.axis[direction].set_visible(False)

x = data2[:, 0]

# Data normalization: https://medium.com/@brijesh_soni/feature-engineering-101-7cb68d293551
# Transmission values
y1 = (data2[:, 1] - np.min(data2[:, 1])) / (np.max(data2[:, 1]) - np.min(data2[:, 1]))
# Get absorption values
y2 = 1 - y1

ax.plot(x, y2, marker = "o")
ax.set_xlabel(r"Temperature, $^{\circ}$C")
ax.set_ylabel("Absorption")
plt.grid(linestyle = "--")
plt.savefig("1_5_absorption_vs_temperature_1.jpg")
plt.show()

fig = plt.figure()
ax = fig.add_subplot(axes_class = AxesZero)

for direction in ["xzero", "yzero"]:
    ax.axis[direction].set_axisline_style("-|>")
    ax.axis[direction].set_visible(False)

for direction in ["left", "bottom"]:
    ax.axis[direction].set_axisline_style("-|>")
    ax.axis[direction].set_visible(True)

for direction in ["right", "top"]:
    ax.axis[direction].set_visible(False)

ax.plot(x, y2)


# Source for the colors of the background
# https://stackoverflow.com/questions/8931268/using-colormaps-to-set-color-of-line-in-matplotlib
# https://saturncloud.io/blog/matplotlib-plot-lines-with-colors-through-colormap-a-guide/
# https://stackoverflow.com/questions/68002782/fill-between-gradient

pts = 1000
xx = np.linspace(0, 45, pts)
cmap = plt.get_cmap("coolwarm")
norm = plt.Normalize(5, xx.max())
colors = cmap(norm(xx))
for i in range(pts):
    ax.axvline(xx[i], color = colors[i], alpha = 0.7)

for i in np.arange(0, xx.max(), 5):
    ax.axvline(i, color = "black", linestyle = "--", alpha = 0.3, linewidth = 1)

for i in np.arange(0, 1.1, 0.2):
    ax.axhline(i, color = "black", linestyle = "--", alpha = 0.3, linewidth = 1)

# ax.fill_between(x, np.zeros_like(y2), y2, color = coolwarm(norm(x)))
ax.plot(x, y2, color = "black", linewidth = 0.8, marker = "o", markerfacecolor = "black")
ax.set_xlabel(r"Temperature, $^{\circ}$C")
ax.set_ylabel("Absorption")
ax.set_xlim(5, xx.max())
plt.savefig("1_5_absorption_vs_temperature_2.jpg")
plt.show()
```
