```python
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import numpy as np
```

```python
dias = np.array([5, 10, 15, 20, 25])         # in mm
powers = np.array([653, 659, 660, 663, 665]) # in microWatts
p0 = 662                                     # in microWatts
losses = - 10 * np.log10(powers / p0)
```

```python
plt.scatter(dias, losses, s = 100, zorder = 2)
plt.xlabel("Curavture Radius [mm]")
plt.ylabel("Loss [dB]")
plt.grid(alpha = 0.4)
plt.savefig("curvature.jpg")
plt.show()
```
