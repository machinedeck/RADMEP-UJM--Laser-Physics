```python
# from https://www.geeksforgeeks.org/python-gaussian-fit/
def gauss(x, a, b):
        y = a * np.exp(-1*b*x**2)
        return y
def gauss_fit(array1, array2, a):
    param, cov = curve_fit(gauss, array1, array2)
    return param[0], param[1]

# For easier data processing
gaussians = [im30cm_v,
        im40cm_v,
        im50cm_v,
        im60cm_v,
        im70cm_v,
        im80cm_v,
        im90cm_v,
        im100cm_v]

# Determine length given that a pixel of the camers is 2.8 x 2.8 um
length = 2.8*len(xx)*1e-4
xxnew = np.linspace(-length/2, length/2, len(xx))

fig = plt.figure(figsize = (15, 8))

ax = fig.add_subplot(axes_class = AxesZero)

# From https://www.tutorialspoint.com/how-to-draw-axis-in-the-middle-of-a-figure-in-matplotlib
ax.spines["bottom"].set_position("center")

for direction in ["xzero", "yzero"]:
    ax.axis[direction].set_axisline_style("-|>")
    ax.axis[direction].set_visible(False)

for direction in ["left", "bottom"]:
    ax.axis[direction].set_axisline_style("-|>")
    ax.axis[direction].set_visible(True)

for direction in ["right", "top"]:
    ax.axis[direction].set_visible(False)

# # From https://www.tutorialspoint.com/how-to-draw-axis-in-the-middle-of-a-figure-in-matplotlib
# ax.spines["bottom"].set_position("center")

lambda_ = 632.8e-9
omega_0_vals = []
for i in range(len(gaussians)):
    z_ = (30 + i*10)*1e-2
    param, cov = curve_fit(gauss, xxnew, gaussians[i])
    omega_0 = (lambda_ * z_ / np.pi) * np.sqrt(param[1] / 2)
    print("- For {} cm\nA: {}\nB: {}\nomega: {}".format(30 + i*10, param[0], param[1], omega_0))
    ax.plot(norm(gauss(xxnew, param[0], param[1]))*8 + 30 + i*10, xxnew, color = "black", linestyle = "--", linewidth = "1.5")
    ax.scatter(norm(gaussians[i])*8 + 30 + i*10, xxnew, color = "red", alpha = 0.5, s = 1.5)
    omega_0_vals.append(omega_0)
# plt.grid()


omega_0_vals = np.array(omega_0_vals)*1e9
print(omega_0_vals.mean())
print(omega_0_vals.std())

omega0 = omega_0_vals.mean()*1e-9

z0 = np.pi * omega0 / lambda_

def width(z):
    return omega0 * np.sqrt(1 + z / z0**2)

xn = np.linspace(0, 101, 101)
plt.plot(xn, width(xn) * 100)
# plt.figure(figsize = (15, 8))

# for i in range(len(gaussians)):
#     param, cov = curve_fit(gauss, xxnew, gaussians[i])
#     print("- For {} cm\nA: {}\nB: {}".format(30 + i*10, param[0], param[1]))
#     plt.plot(xxnew, norm(gauss(xnew, param[0], param[1]))*8 + 30 + 30 + i*10)
#     plt.plot(xxnew, norm(im30cm_v)*8 + 30 + i*10, color = "orange", linewidth = 0.5)
# plt.show()
```
