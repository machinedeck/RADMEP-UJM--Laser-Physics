```python
import numpy as np
```

```python
L = 28.5 + 15 # in cm
R = 60 # in cm
wav = 632.8e-7 # wavelength in cm

omega0 = ((wav**2 / np.pi**2) * (L / 2) * (R - L / 2))**(1/4)
```

```python
print(omega0*10) # in mm
```

```python
wav_ = 632.8e-9
omega0_ = omega0 * 1e-3
def w(z):
    z0 = np.pi * omega0_**2 / wav_
    return omega0_ * np.sqrt(1 + (z / z0)**2)
```

```python
f = 8e-3

phi = 4 * wav_ * f / (np.pi * omega0_)
print(phi)
```
